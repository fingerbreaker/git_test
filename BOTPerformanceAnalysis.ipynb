### Cleaning

```python
df_master_funnel['Funnel Prod Phone'] = df_master_funnel['Funnel Prod Phone'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df_freshchat['User properties Phone number'] = df_freshchat['User properties Phone number'].astype(str).str.replace(r'\.0$', '', regex=True)
```

### Merging the Freshchat and Funnel DF

```python
merged_df = df_freshchat.merge(
    df_master_funnel,
    left_on='User properties Phone number',
    right_on='Funnel Prod Phone',
    how='inner'
)
merged_df['Last Txn Date'] = pd.to_datetime(merged_df['Funnel Prod Last Txn Date'])
merged_df['Initiated At'] = pd.to_datetime(merged_df['Initiated At'])

merged_df['Funnel Prod Phone'].nunique()
merged_df
```

```python
merged_df.groupby(['Funnel Prod Last Funnel Stage Completed'], sort=True)['Funnel Prod Phone'].count().reset_index().sort_values(by='Funnel Prod Phone', ascending=False)
```

### Checking for txn less than 4 hr conversions

```python
filtered_txn_less_4 = merged_df[merged_df['Disposition'] == 'Transaction_Pending_Less_4hr']
filtered_txn_less_4_count = filtered_txn_less_4['Funnel Prod Phone'].nunique()
```

```python
matched_users_txn_less_4 = filtered_txn_less_4[filtered_txn_less_4['Funnel Prod Last Txn Date'] > filtered_txn_less_4['Initiated At']]

count_txn_less_4 = matched_users_txn_less_4['Funnel Prod Phone'].nunique()

print(f"Number of queries resolved: {count_txn_less_4}\nNumber of queries not resolved: {filtered_txn_less_4_count - count_txn_less_4}\nPercentage not solved for: {(filtered_txn_less_4_count - count_txn_less_4)/filtered_txn_less_4_count*100}%")
print
```
